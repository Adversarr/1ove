```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import *
```

```python
f = lambda x,y,z: (x**2 + 9/4 * y**2 + z ** 2 - 1)**3 -(x**2*z**3+9*y**2*z**3/80)
```

```python
phi = np.linspace(0, np.pi, 50)
theta = np.linspace(-np.pi, np.pi, 100)
phi, theta = np.meshgrid(phi, theta)
r = 1.5
x, y, z = r * np.sin(phi) * np.cos(theta), r * np.sin(phi) * np.sin(theta),  r * np.cos(phi) + 1
shape = x.shape
```

```python

xl, yl, zl = [], [], []
for xi, yi, zi in zip(x.reshape(-1), y.reshape(-1), z.reshape(-1)):
    x0 = np.array([xi, yi, zi])
    x_ = fmin(lambda x: np.abs(f(x[0], x[1], x[2])), x0, disp=False)
    if np.abs(f(x_[0], x_[1], x_[2])) < 1e-4 or True:
        xl.append(x_[0])
        yl.append(x_[1])
        zl.append(x_[2])
xl = np.array(xl).reshape(shape)
yl = np.array(yl).reshape(shape)
zl = np.array(zl).reshape(shape)
fig = plt.figure(figsize=(10, 12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xl, yl, zl)
```
